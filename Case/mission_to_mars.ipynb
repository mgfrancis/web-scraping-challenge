#NASA MARS NEWS

```python
# Dependencies
from bs4 import BeautifulSoup as bs
import requests
```

```python
# URL of NASA MARS NEWS to be scraped
url = 'https://mars.nasa.gov/news/'
```

```python
# Retrieve page with the requests module
response = requests.get(url)
response.text
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = bs(response.text, 'html.parser')
```

```python
# Examine the results, then determine element that contains sought info
print(soup.prettify())
```

```python
# results are returned as an iterable list
results = soup.find_all('div', class_="slide")
len(results)
```

```python
#check if results contain what we need
print(results[1].prettify())
```

```python
# Loop through returned results
for result in results:
    # Error handling
    try:
        # Identify and return news titles
        news_title = result.find('div', class_='content_title').text
        # Identify and return news teasers 
        news_p = result.find('div', class_='rollover_description_inner').text
        

        # Print results only if title, price, and link are available
        if (news_title and news_p):
            print(news_title)
            print(news_p)
            print('-------------')
    except Exception as e:
        print(e)
```

#Mars Hemispheres

```python
# URL of NASA MARS NEWS to be scraped
url = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
```

```python
# Retrieve page with the requests module
response = requests.get(url)
response.text
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = bs(response.text, 'html.parser')
```

```python
# Examine the results, then determine element that contains sought info
print(soup.prettify())
```

```python
# results are returned as an iterable list
results = soup.find_all('div', class_="item")
len(results)
```

```python
#check if results contain what we need
print(results[0].prettify())
```

```python
# Loop through returned results

img_url = []
title = []
hemisphere_image_urls = []
for result in results:
    # Error handling
    try:
        # Identify and return news titles
        h3 = result.find('h3').text
        # Identify and return news teasers 
        img = result.find('img', class_='thumb')['src']
        

        # Print results only if title, and url are available
        if (h3 and img):
            title.append(h3)
            img_url.append(img)
            hemisphere_image_urls.append({'title':h3, 'img_url': img})
        
    except Exception as e:
        print(e)
    
hemisphere_image_urls
```
